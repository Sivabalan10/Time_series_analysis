## Handling missing values

```python
df_filled = df.fillna(method='ffill')
```

```python
df_filled.isna().sum()
```

```python
df_filled
```

## Remove Duplicates

```python
df_filled.shape
```

```python
data_cleaned = df_filled.drop_duplicates(ignore_index=True)
```

```python
data_cleaned.shape
```

```python
data_cleaned
```

## Outlier Removal

```python
from scipy.stats import zscore
```

```python
z_scores = zscore(data_cleaned['temperature'])
outliers = abs(z_scores) > 3
outliers
```
